```python
from time import sleep as s
from os import system, name

player_name = None
purpose = None


def clear(): 
  
    if name == 'nt': 
        _ = system('cls') 
  
    # for mac and linux(here, os.name is 'posix') 
    else: 
        _ = system('clear') 
        

def intro():
	clear()
	global player_name
	global purpose
	player_name = input ("Welcome to the ministry of magic. Please state your name ")
	purpose = input (f"Welcome {player_name.title()}. Please state your purpose at the ministry ")
	print(f"Oh, and I forgot the number code. But because you are {player_name.title()}, I suppose you already know the code is 62442.")
	code = input ("What is the code, then? ")
	if code == "62442":
		print(f"Correct, {player_name.title()}. You may enter.")
	else:
		print(f"Wrong, {player_name.title()}. Try again.")
		intro()
	print("An elevater takes you down, down, down to the ministry")
	s(2)
	hall()

def hall():
	clear()
	print("You are now in a great big hall with a fountain with statues made of gold.")
	print ("There are lot's of halls that branch off from the big area.")
	hall=input("You could also go to the front desk. d for desk, h for hallway 1, and a for hallway 2 ")
	if hall == 'd':
		print("You choose to go to the front desk to sign in") 
		desk()
	elif hall == 'h':
		print("You slip down hallway 1")
		hall_1()
	elif hall == 'a':
		print ("You walk down hallway 2")
		hall_2()
def desk():
	print("You go to the desk")
	print(f'There is a woman behind the desk. "Why are you here, {player_name.title()}?"')
	print(f'You respond with what you told the robot earlier "{purpose}" you respond')
	print(f'"{purpose}, huh." she responds. "Our {purpose} department is just down that hallway. Hallway 1."')
	purpose_way = input("Would you like to go the way the woman points you? yes/no ")
	if purpose_way == 'yes':
		print("Good choice.")
		hall_1()
	else:
		print("Oh well.")
		so= input ("Leave the ministry, take hallway 2, or go back to the desk? l for leave, t for hallway 2, and d for desk ")
		if so == 'l':
			print("You leave the ministry.")
			print("game over.")
		if so == 't':
			("You take the other hallway? Alright then.")
			hall_2()
		if so == 'd':
			print("Back to the front desk!")
			desk()
def hall_1():
	print("You are in the hallway")
	department= input (f"You see a door that says {purpose} department! Will you go in? y/n ")
	if department == 'y':
		print("You walk in, and are immedietly greeted by a robot!")
		print(f"He is very nice, and helps you throughout the whole day! He gives you {purpose}!")
		print("At the end of the day, you get a very nice big glass of butterbeer!")
	
def hall_2():
	print("You are in the hallway")
	print("Suddenly, you feel something wet on your face!")
	print("You look up and see...a glass of butterbeer pouring butterbeer on you!")
	print ("It falls onto the ground, and you slip onto your face!")
	butterbeer = input ("Do you lay on the ground and don't go anywhere, or run back to the desk? l for lay, d for desk ")
	if butterbeer == 'l':
		print("You lay there sulking.")
		print ("game over.")
	else:
		print ("You rush back to the desk!")
		desk()
	
intro()
```

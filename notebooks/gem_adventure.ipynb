```python
from time import sleep as s
from os import system, name 
from random import randint as r

def clear(): 
  
    if name == 'nt': 
        _ = system('cls') 
  
    # for mac and linux(here, os.name is 'posix') 
    else: 
        _ = system('clear') 
                                              
print("""


   ___                                        
  / __|___ _ __                               
 | (_ / -_) '  \                              
  \___\___|_|_|_|            _                
       /_\  __| |_ _____ _ _| |_ _  _ _ _ ___ 
      / _ \/ _` \ V / -_) ' \  _| || | '_/ -_)
     /_/ \_\__,_|\_/\___|_||_\__|\_,_|_| \___|
                                                       
                                                       
""")


def intro():
	s(3)
	clear()
	print("Welcome to the land, adventurer!")
	s(2)
	print("I take it you are on an adventure to find the long lost gem?")
	s(2)
	print("Well, let me give you a hint. The lizard on the desert path can help you to find it.")
	s(2)
	print("Would you like to find the lizard and go through the desert path?")
	s(2)
	choice = input("yes/no ")
	if choice == 'yes':
		print("You walk along quickly")
		desert()
	elif choice == 'no':
		("You decide the lizard probably won't help you, and you walk back.")
		intro()
	else:
		print("Please enter a valid answer.")
		s(2)
		intro()
def desert():
    clear()
    print ("You walk along, trying to get through the desert, as you are very hot")
    s(2)
    print("Suddenly, a lizard walks in front of you.")
    s(2)
    print('"Are you looking for the gem?" he asks')
    s(2)
    print('"I will tell you how you can get the gem. First, you must guess which one of my paws holds the money."')
    lizard= r(1,2)
    choice= input ("Guess in which 'hand' the lizard holds the money. 1 for right, 2 for left ")
    choice = int(choice)
    if choice == lizard:
        print('"Good job, traveler! You get the money!"')
        s(2)
        print("Lizard money seems to be an assortment of pepperoni and olives")
    else:
        print('"Ah, well, alas, traveler,you did not guess. But, I suppose I can spare just a few of my olives..." says the lizard')
        s(2)
        print('And so, the lizard gives you a small handful of olives and pepperoni. "Use them wisely" he says')

    moving_on = input("Would you like to move on your quest? q for move on, s to restart ")

    if moving_on == 'q':
        castle_start()
    elif moving_on == 's':
        intro()
    else:
        print("Sorry, that's not a valid answer. I guess you'll need redo this")
        s(2)
        desert()
    

def castle_start():
	clear()
	print("You walk away from the desert, smiling, as you now have some olives and pepperoni")
	s(2)
	print("When just before you--you see a giant castle. Would you like to enter the castle, or run back to the lizard?")
	choices = input("c to enter the castle, l for returning to the lizard.")
	if  choices == 'c':
		print ("You  move on into the big castle ahead of you. Hopefully the gem will be there.")
		s(4)
		castle()
	elif choices == 'l':
		("You decide the castle is way to intimadating and you rush back to the lizard. Maybe he can help you with your fears")
		s(2)
		desert()
	else:
		("Not a valid answer. Please try again.")
		s(2)
		castle_start()

def castle():
	clear()
	print("You rush into the castle, however intimadating it might be, it's no match for YOU!")
	s(2)
	print("You are faced already with a fork in the halls however, although you just got in.")
	hmm = input("Do you want to take the right or the left fork? l for left, r for right.")
	if hmm == 'l':
		print("You take the left fork. Hopefully it was a good one.")
		s(2)
		move()
	elif hmm == 'r':
		print("You decide on the right one. Hopefully it will lead you to the gem.")
		s(2)
		move()
def move():
	clear()
	print("You continue down the hall, unsure if it was the right one, but not willing to turn back.")
	s(2)
	print("Suddenly, the passageway comes to an end. There's a large door.")
	door = input("Open the door, or turn around and walk down the hall? o for open the door, and w for walking down the hall.")
	if door == 'o':
		gem()
	elif door == 'w':
		print("You decide not to open the door, and you turn around.")
		s(2)
		move()
	else:
		("Please enter a valid answer.")
		s(4)
		move()
def gem():
	print("The door creaks open. You find yourself in bright sunlight.")
	s(2)
	print("You seem to be in a garden. You must've walked the whole castle through in that hall!")
	s(2)
	print("You see a sparkling light blue gem. It seems to be either a sapphire or an aquamarine.")
	s(2)
	print("You reach out to touch it. This was all your quest was for. When...")
	s(2)
	old_lady()

def old_lady():
	s(2)
	clear()
	print("An old lady steps in front of you!")
	s(2)
	print('"Dont you be taking my stone yet!" she says')
	s(2)
	print('"You go making me my pizza first! I like pepperoni and olives, now get to work!"')
	s(2)
	print("Thankfully, you remember the lizard money that the lizard gave you.")
	s(2)
	print("The lady givs you ingredients for the dough, and mozzarella cheese.")
	s(2)
	print("Very much later...")
	s(3)
	print("You give the pizza to the lady. She eats it all.")
	s(2)
	print('"Hmm...good pizza" she says. She lets you pass and get the gem.')
	s(2)
	print("A quest well done, adventurer!")    

intro ()
    
```
